### Basic array operations
- algebraic operations 
- logical operations

```python
a = np.array([[-4,6,8], [9, 3, 1], [1, 8, 5]])
b = np.array([[11,16,18], [90, 37, 81], [61, 48, 75]])
```

```python
# Write a NumPy function to add, subtract, multiply and divide arrays 'a' and 'b'
def add_Arr():
  arr = np.add(a,b)
  print(arr)
  arr = np.subtract(a,b)
  print(arr)
  arr = np.multiply(a,b)
  print(arr)
  arr = np.divide(a,b)
  print(arr)
add_Arr()
```

```python
# Write a NumPy function to get the sum of the array 'a' along axis=0 
def add_Arr_x():
  arr = np.sum(a,axis=0)
  print(arr)  
add_Arr_x()
# Write a NumPy function to get the sum of the array 'a' along axis=1 
def add_Arr_y():
  arr = np.sum(a,axis = 1)
  print(arr)  
add_Arr_y()
```

```python
# Write a NumPy function to reverse the array 'a' 
def reverse_func():
  rev = np.flip(a)
  print(rev)
reverse_func()

# Write a NumPy function to reverse the array 'a' along axis=0
def reverse_func0():
  rev = np.flip(a,axis = 0)
  print(rev)
reverse_func0()

# Write a NumPy function to reverse the array 'a' along axis=1
def reverse_func1():
  rev = np.flip(a,axis = 1)
  print(rev)
reverse_func1()
```

- greater than equal to 5 –> ‘large’

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = np.genfromtxt(url, delimiter=',', dtype='float')
names = ('sepallength', 'sepalwidth', 'petallength', 'petalwidth', 'species')

# If the dtype was object the elements were byte strings

def bin_func():
  lst  = []
  for i in range(150):
    if iris[i][2] < 3:
      lst.append('small')
    if iris[i][2] >= 3 and int(iris[i][2]) < 5  :
      lst.append('medium')
    if iris[i][2] >= 5:
      lst.append('large')
  print(lst)
bin_func()

```

Q6 Write a Pandas function to find the most frequent value of petal length (3rd column) in iris dataset

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = np.genfromtxt(url, delimiter=',', dtype='float')
names = ('sepallength', 'sepalwidth', 'petallength', 'petalwidth', 'species')

def freq():
  df = pd.DataFrame(iris)
  df = df.set_axis(names, axis=1)
  ar = df.mode()
  print(ar['petallength'][0])
freq()
```
