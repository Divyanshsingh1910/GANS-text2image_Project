```python
# Write a Pandas function to drop rows that contain a missing value. 
def drop_missing_rows(df):
    df.dropna(inplace=True)
    return df
```

```python
# Write a Pandas function to find the correlation between the first two columns.
def find_correlation(df):
    df_subset = df.iloc[:, :2]
    correlation = df_subset.corr()
    return correlation.iloc[0,1]
```

```python
# Write a Pandas function to find the mean, median and standard deviation of column 3.
def find_stats(df):
    col_3 = df.iloc[:, 2]
    mean = col_3.mean()
    median = col_3.median()
    std_dv = col_3,std()
    return {'mean': mean, 'median': median, 'standard deviation': std_dv}
```

# Q5 Write a Pandas function to bin the petal length (3rd) column of iris_2d to form a text array, such that if petal length is:
- Less than 3 –> ‘small’
- 3-5 –> ‘medium’
- greater than equal to 5 –> ‘large’

```python
def bin_petal_length(iris_2d):
    bins = [0, 3, 5, float("inf")]
    labels = ["small", "medium", "large"]
    iris_2d['petal_length_category'] = pd.cut(iris_2d['petal_length'], bins=bins, labels=labels)
    return iris_2d
```
