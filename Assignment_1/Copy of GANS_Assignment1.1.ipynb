# Q5 Write a Pandas function to bin the petal length (3rd) column of iris_2d to form a text array, such that if petal length is:
- Less than 3 –> ‘small’
- 3-5 –> ‘medium’
- greater than equal to 5 –> ‘large’

```python
bin = np.digitize(iris[:, 2].astype('float'), [0, 3, 5])
category = {1: 'small', 2: 'medium', 3: 'large'}
text_arr = [category[x] for x in bin]
print(text_arr)
```
